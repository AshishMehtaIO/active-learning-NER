```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set()
```

```python
dir = os.getcwd()+'/Results/ESP/'
labels = ['Uncertainity Least Confidence', 'Uncertainity Total Entropy', 'Baseline Max length', 'QBC 3 Learners', 'Baseline Random']
for iter, filename in enumerate(sorted(os.listdir(dir))):
    print(filename)
    data = np.loadtxt(dir+filename)
    x = data[:,0]
    y = data[:,1]
    plt.plot(x,y, label=labels[iter])
    plt.legend()
    plt.ylabel('F1 score')
    plt.xlabel('Number of sentences in training set')
    plt.title('Learning Curve for Active Learning starategies on ESP CONLL2002 dataset')
    plt.savefig('ESP.png', dpi=600)
```

```python
dir = os.getcwd()+'/Results/ESP/'
labels = ['Uncertainity Least Confidence', 'Uncertainity Total Entropy', 'Baseline Max length', 'QBC 3 Learners', 'Baseline Random']

threshold = 0.7

for iter, filename in enumerate(sorted(os.listdir(dir))):
    x=[]
    y=[]
    print(filename)
    data = np.loadtxt(dir+filename)
    for d in data:
        if d[1] < threshold:
            x.append(d[0])
            y.append(d[1])
    plt.plot(x,y, label=labels[iter])
    plt.legend()
    plt.ylabel('F1 score')
    plt.xlabel('Number of sentences in training set')
    plt.title('Learning Curve for Active Learning starategies on ESP CONLL2002 dataset')
    plt.savefig('ESP_thresh.png', dpi=600)
```
